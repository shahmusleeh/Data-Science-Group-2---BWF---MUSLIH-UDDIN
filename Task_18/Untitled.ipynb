# Assigned Task

### Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('tips.csv')
df
```

### Set the style

```python
sns.set_style("whitegrid")
```

### Create a scatter plot

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='time')
```

### Add title and labels

```python
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

### Show plot

```python
plt.show()
```

### Now Running the code at once we will get the following output

```python
sns.set_style("whitegrid")

sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='time')

plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```

### Line Plot

#### Here we will do the same as we did for scatter plot and will run the code as follows

```python
sns.set_style("darkgrid")

sns.lineplot(data=df, x='size', y='total_bill', hue='time')

plt.title('Line Plot of Total Bill by Size')
plt.xlabel('Size')
plt.ylabel('Total Bill')

plt.show()
```

### Bar Plot

```python
sns.set_style("ticks")

sns.barplot(data=df, x='day', y='total_bill', hue='sex', palette='muted')

plt.title('Bar Plot of Total Bill by Day and Sex')
plt.xlabel('Day')
plt.ylabel('Total Bill')

plt.show()
```

### Count Plot

```python
sns.set_style("white")

sns.countplot(data=df, x='day', hue='sex', palette='coolwarm')

plt.title('Count Plot of Days by Sex')
plt.xlabel('Day')
plt.ylabel('Count')

plt.show()
```

### Box Plot

```python
sns.set_style("dark")

sns.boxplot(data=df, x='day', y='total_bill', hue='sex', palette='Set3')

plt.title('Box Plot of Total Bill by Day and Sex')
plt.xlabel('Day')
plt.ylabel('Total Bill')

plt.show()
```

# Introduction to Matplotlib (basic plots: line, scatter, bar)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ("IMDb_Dataset.csv")
```

### Line Plot

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['IMDb Rating'], label='IMDb Rating')
plt.plot(df['Year'], df['MetaScore'], label='MetaScore', linestyle='--', marker='o')
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('IMDb Rating and MetaScore Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['IMDb Rating'], df['MetaScore'], color='r', label='Movies')
plt.xlabel('IMDb Rating')
plt.ylabel('MetaScore')
plt.title('Scatter Plot: IMDb Rating vs MetaScore')
plt.legend()
plt.grid(True)
plt.show()
```

### Bar Chart

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Title'], df['IMDb Rating'], color='skyblue', edgecolor='blue')
plt.xlabel('Movie Title')
plt.ylabel('IMDb Rating')
plt.title('Vertical Bar Chart: IMDb Rating for Each Movie')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Pie Chart

```python
genre_counts = df['Genre'].value_counts()
colors = ['lightcoral', 'yellowgreen', 'lightskyblue', 'orange', 'gold', 'purple']
explode = [0.1] + [0] * (len(genre_counts) - 1)

plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', colors=colors[:len(genre_counts)], explode=explode)
plt.title('Pie Chart: Proportion of Different Genres')
plt.legend(genre_counts.index, loc='upper left')
plt.show()
```

### Customize Line Styles and Colors

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['IMDb Rating'], label='IMDb Rating', marker='o', linestyle='-', color='blue')
plt.plot(df['Year'], df['MetaScore'], label='MetaScore', marker='x', linestyle='--', color='red')
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Customized IMDb Rating and MetaScore Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

### Create a Basic Scatter Plot

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['IMDb Rating'], df['MetaScore'])
plt.xlabel('IMDb Rating')
plt.ylabel('MetaScore')
plt.title('Scatter Plot: IMDb Rating vs MetaScore')
plt.grid(True)
plt.show()
```

### Customize Marker Styles, Sizes, and Colors

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['IMDb Rating'], df['MetaScore'], marker='o', s=100, c='red', alpha=0.5)
plt.xlabel('IMDb Rating')
plt.ylabel('MetaScore')
plt.title('Customized Scatter Plot: IMDb Rating vs MetaScore')
plt.grid(True)
plt.show()
```

### Plot Multiple Scatter Plots on the Same Graph

```python
colors = {'Action': 'red', 'Drama': 'blue', 'Comedy': 'green', 'Thriller': 'purple', 'Adventure': 'orange', 'Animation': 'brown', 'Crime': 'black', 'Biography': 'white', 'Horror': 'yellow', 'Fantasy': 'pink'}
plt.figure(figsize=(10, 5))
for genre in df['Genre'].unique():
    subset = df[df['Genre'] == genre]
    plt.scatter(subset['IMDb Rating'], subset['MetaScore'], label=genre, color=colors[genre])
plt.xlabel('IMDb Rating')
plt.ylabel('MetaScore')
plt.title('Scatter Plot: IMDb Rating vs MetaScore by Genre')
plt.legend()
plt.grid(True)
plt.show()
```

### Add Labels and Titles to the Scatter Plot

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['IMDb Rating'], df['MetaScore'], marker='o', s=100, c='blue', alpha=0.5)
for i in range(len(df)):
    plt.text(df['IMDb Rating'][i], df['MetaScore'][i], df['Title'][i], fontsize=9)
plt.xlabel('IMDb Rating')
plt.ylabel('MetaScore')
plt.title('Scatter Plot with Labels: IMDb Rating vs MetaScore')
plt.grid(True)
plt.show()
```

# Mini Project

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

df.head()
```

```python
import matplotlib.pyplot as plt

age_counts = df['Age'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
plt.plot(age_counts.index, age_counts.values, label='Number of Passengers', marker='o')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers Over Ages')
plt.grid(True)
plt.legend()
plt.show()
```

```python
age_survival = df.groupby('Age')['Survived'].mean()
plt.figure(figsize=(10, 5))
plt.plot(age_survival.index, age_survival.values, label='Survival Rate', marker='x')
plt.xlabel('Age')
plt.ylabel('Survival Rate')
plt.title('Survival Rate Over Ages')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
colors = {0: 'red', 1: 'blue'}
plt.scatter(df['Age'], df['Fare'], c=df['Survived'].map(colors), alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot: Fare vs Age')
plt.grid(True)
plt.show()
```

```python
class_survival = df.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(10, 5))
plt.bar(class_survival.index, class_survival.values, color=['skyblue', 'orange', 'green'])
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Class')
plt.grid(True)
plt.show()
```

```python
gender_survival = df.groupby('Sex')['Survived'].mean()
plt.figure(figsize=(10, 5))
plt.bar(gender_survival.index, gender_survival.values, color=['purple', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender')
plt.grid(True)
plt.show()
```

```python
survival_counts = df['Survived'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(survival_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%', colors=['lightcoral', 'yellowgreen'], explode=[0.1, 0])
plt.title('Proportion of Survivors')
plt.show()
```

# Data Manipulation in Pandas

# Group the DataFrame by a specific column and calculate the mean of each group

```python
df.groupby ('gender').agg({'math score': ["mean"]})
```

# Group the DataFrame by multiple columns and calculate the sum of each group.

```python
df.groupby ('gender').agg({'reading score': ["sum"]})
```

```python
df.groupby ('gender').agg({'reading score': ["mean","sum"]})
```

```python
df.groupby ('gender').agg({'reading score': ["mean","sum"],
                          'math score': ["mean","sum"]})
```

# Calculate the size of each group

```python
df.groupby('gender').size()
```

# Select rows based on multiple conditions

```python
df[(df['math score'] > 50) & (df['reading score'] > 80)]
```

# Use isin to filter rows based on a list of values

```python
df[df['gender'].isin(['male', 'female'])]
```

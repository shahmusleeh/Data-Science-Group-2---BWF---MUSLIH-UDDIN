# Advanced plotting techniques

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tested.csv')
```

```python
df
```

### Histogram to understand the Distribution

```python
import warnings
warnings.filterwarnings("ignore")
for i in df.select_dtypes(include = "number").columns:
    sns.histplot(data = df, x = i)
    plt.show()
```

### Generate a figure with 2x2 subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
```

### Line plot

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs[0, 0].plot(df['PassengerId'], df['Fare'], label='Fare')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('PassengerId')
axs[0, 0].set_ylabel('Fare')
axs[0, 0].legend()
```

### Scatter plot

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs[0, 1].scatter(df['Age'], df['Fare'], c='blue', label='Fare vs Age')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('Fare')
axs[0, 1].legend()
```

### Bar Plot

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
df['Pclass'].value_counts().plot(kind='bar', ax=axs[1, 0])
axs[1, 0].set_title('Bar Plot')
axs[1, 0].set_xlabel('Pclass')
axs[1, 0].set_ylabel('Count')
```

### Histogram

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs[1, 1].hist(df['Age'].dropna(), bins=20, color='green', alpha=0.7)
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Create a histogram for a given numerical data set

```python
plt.hist(df['Fare'], bins=30, color='purple', alpha=0.7)
plt.title('Fare Histogram')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

### Plot multiple histograms on the same figure

```python
plt.hist(df['Age'].dropna(), bins=20, alpha=0.5, label='Age', color='blue')
plt.hist(df['Fare'], bins=20, alpha=0.5, label='Fare', color='red')
plt.title('Age and Fare Histograms')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### Normalize the histograms

```python
plt.hist(df['Age'].dropna(), bins=20, alpha=0.5, density=True, label='Age', color='blue')
plt.hist(df['Fare'], bins=20, alpha=0.5, density=True, label='Fare', color='red')
plt.title('Normalized Age and Fare Histograms')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### Create a box plot for a given numerical data set

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['Fare'])
plt.title('Box Plot of Fare')
plt.ylabel('Fare')
plt.show()
```

### Generate multiple box plots in a single figure to compare different data sets

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title('Box Plot of Age and Fare')
plt.ylabel('Value')
plt.show()
```

### Customize the box plots with titles, labels, and change the colors

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[['Age', 'Fare']], palette="Set2")
plt.title('Customized Box Plot of Age and Fare')
plt.ylabel('Value')
plt.show()
```
